```python
#Trying on another dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist();
```

```python
heart_disease.plot.hist(figsize = (10,30), subplots = True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# 1.pyplot method
over_50.plot(kind = 'scatter', x = 'age', y = 'chol', c = 'target');
```

```python
import matplotlib.pyplot as plt
# 2.Object oriented method
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax);
ax.set_xlim([45,100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                  y = over_50["chol"],
                  c = over_50["target"])

# Customize the plots
ax.set(title = "Heart Disease and Cholesterol levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title = "target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = "--");
```

```python
over_50.head()
```

```python
# Subplot of chol, age and thalach
import matplotlib.pyplot as plt
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize = (10, 10),
                               sharex = True)

# Add data to ax0
scatter0 = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])

# Naming the plot ax0
ax0.set(title = "Age vs Cholesterol Plot",
        ylabel = "Cholesterol");

# Adding a legend to ax0
ax0.legend(*scatter.legend_elements(), title = "target");

# Adding a horizontal line to the plot
ax0.axhline(y = over_50["chol"].mean(),
            linestyle = "--");



# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                   y = over_50["thalach"],
                   c = over_50["target"])

# Naming the plot ax1
ax1.set(title = "Age vs Thalach Plot",  # Thalach means max heart rate
        xlabel = "Age",
        ylabel = "Thalach");

# Adding a Legend to ax1
ax1.legend(*scatter.legend_elements(), title = "target");

# Adding a horizontal line to the plot
ax1.axhline(y = over_50["thalach"].mean(),
            linestyle = "--");
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                  y = over_50["chol"],
                  c = over_50["target"],
                  cmap = "winter") # This changes the scheme

# Customize the plots
ax.set(title = "Heart Disease and Cholesterol levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title = "target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = "--");
```

```python
# Customizing the y and x axis limitations

# Subplot of chol, age and thalach
import matplotlib.pyplot as plt
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize = (10, 10),
                               sharex = True)

# Add data to ax0
scatter0 = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"],
                      cmap = "winter")

# Naming the plot ax0
ax0.set(title = "Heart Disease and Cholesterol Levels",
        ylabel = "Cholesterol");

# Changing the x axis limits
ax0.set_xlim([50, 80])

# Adding a legend to ax0
ax0.legend(*scatter.legend_elements(), title = "target");

# Adding a horizontal line to the plot
ax0.axhline(y = over_50["chol"].mean(),
            linestyle = "--");



# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                   y = over_50["thalach"],
                   c = over_50["target"],
                   cmap = "winter")

# Naming the plot ax1
ax1.set(title = "Heart Disease and Max Heart Rate",  
        xlabel = "Age",
        ylabel = "Thalach"); # Thalach means max heart rate

# Changing ax1 x axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Adding a Legend to ax1
ax1.legend(*scatter.legend_elements(), title = "target");

# Adding a horizontal line to the plot
ax1.axhline(y = over_50["thalach"].mean(),
            linestyle = "--");
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
